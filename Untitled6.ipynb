###ขั้นที่ 1 เชื่อมตาราง Products และ Sales_Order_Items ด้วย PRODUCTID เพื่อเพิ่ม column : PRODCATEGORYID (ชนิดสินค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['PRODUCT_C'] = Sales_Order_Items ['PRODUCTID'].map(Products.set_index('PRODUCTID')['PRODCATEGORYID']) 
Sales_Order_Items
```

###ขั้นที่ 2 เชื่อมตาราง Sales_Orders และ Sales_Order_Items ด้วย SALESORDERID เพื่อเพิ่ม column : PARTNERID (คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['PARTNERID'] = Sales_Order_Items ['SALESORDERID'].map(Sales_Orders.set_index('SALESORDERID')['PARTNERID']) 
Sales_Order_Items
```

###ขั้นที่ 3 เชื่อมตาราง Sales_Orders และ BusinessPartners ด้วย PARTNERID เพื่อเพิ่ม column : ADDRESSID (ที่อยู่คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['ADDRESS_ID'] = Sales_Order_Items ['PARTNERID'].map(BusinessPartners.set_index('PARTNERID')['ADDRESSID']) 
Sales_Order_Items
```

###ขั้นที่ 4 เชื่อมตาราง Sales_Orders และ Addresses ด้วย ADDRESSID เพื่อเพิ่ม column : REGION (ทวีปที่อยู่คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['REGION'] = Sales_Order_Items ['ADDRESSID'].map(Addresses .set_index('ADDRESSID')['REGION','']) 
Sales_Order_Items
```
