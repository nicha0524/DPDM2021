#โจทย์คำถามข้อ 1 
         ทวีปไหน เมืองอะไร ที่มีการสั่งชื้อจักยานมากที่สุด และจักรยานชนิดใดถูกสั่งซื้อเยอะที่สุดเรียงจากมากไปน้อย
    

###ขั้นที่ 1 เชื่อมตาราง Products และ Sales_Order_Items ด้วย PRODUCTID เพื่อเพิ่ม column : PRODCATEGORYID (ชนิดสินค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['PRODUCT_C'] = Sales_Order_Items ['PRODUCTID'].map(Products.set_index('PRODUCTID')['PRODCATEGORYID']) 
Sales_Order_Items
```

###ขั้นที่ 2 เชื่อมตาราง Sales_Order_Items และ Sales_Order ด้วย SALESORDERID เพื่อเพิ่ม column : PARTNERID (คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['PARTNERID'] = Sales_Order_Items ['SALESORDERID'].map(Sales_Orders.set_index('SALESORDERID')['PARTNERID']) 
Sales_Order_Items
```

###ขั้นที่ 3 เชื่อมตาราง Sales_Order_Items และ BusinessPartners ด้วย PARTNERID เพื่อเพิ่ม column : ADDRESSID (ที่อยู่คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['ADDRESS_ID'] = Sales_Order_Items ['PARTNERID'].map(BusinessPartners.set_index('PARTNERID')['ADDRESSID']) 
Sales_Order_Items
```

###ขั้นที่ 4 เชื่อมตาราง Sales_Order_Itemsและ Addresses ด้วย ADDRESSID เพื่อเพิ่ม column : REGION (ทวีปที่อยู่คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['REGION'] = Sales_Order_Items ['ADDRESS_ID'].map(Addresses .set_index('ADDRESSID')['REGION']) 
Sales_Order_Items
```

ขั้นที่ 4 เชื่อมตาราง Sales_Order_Items และ Addresses ด้วย ADDRESSID เพื่อเพิ่ม column : CITY (เมืองที่อยู่คู่ค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['CITY'] = Sales_Order_Items ['ADDRESS_ID'].map(Addresses .set_index('ADDRESSID')['CITY']) 
Sales_Order_Items
```

ขั้นที่ 4 เชื่อมตาราง Sales_Order_Items และ ProductCategory ด้วย PRODCATEGORYID และ PRODUCT_C	 เพื่อเพิ่ม column : 	SHORT_DESCR (ชื่อเต็มของชนิดสินค้า) โดยใช้ตาราง Sales_Order_Items เป็นตารางหลัก

```python
Sales_Order_Items['SHORT_DESCR'] = Sales_Order_Items ['PRODUCT_C'].map(ProductCategory .set_index('PRODCATEGORYID')['SHORT_DESCR']) 
Sales_Order_Items
```

```python
set(Sales_Order_Items['REGION'])
```

```python
Sales_Order_Items['Region_full'] = Sales_Order_Items ['REGION'].map(region_fall .set_index('Region')['Region_full_name']) 
Sales_Order_Items
```

### Split data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5, random_state=6)
X_traintrain, X_val, y_traintrain, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=6) 
```

### KNN
Import 

```python
from sklearn.neighbors import KNeighborsClassifier
```

###Define

####Train Model 1

```python
neigh1 = KNeighborsClassifier(n_neighbors=1) #เชื่อใกล้สุดแค่คนเดียว
neigh1.fit(X_traintrain,y_traintrain)
```

####Train Model 2

```python

neigh2 = KNeighborsClassifier(n_neighbors=3,weights='distance') #เชื่อคนใกล้มากว่าคนไกล
neigh2.fit(X_traintrain,y_traintrain)
```

####Train Model 3

```python

neigh3 = KNeighborsClassifier(n_neighbors=5,weights='distance') #เชื่อทุกคนเท่ากัน
neigh3.fit(X_traintrain,y_traintrain)
```

####Train Model 4

```python

neigh4 = KNeighborsClassifier(n_neighbors=9,weights='distance') #เชื่อคนใกล้มากว่าคนไกล
neigh4.fit(X_traintrain,y_traintrain)
```

####Train Model 5

```python

neigh5 = KNeighborsClassifier(n_neighbors=11,weights='distance') #เชื่อคนใกล้มากว่าคนไกล
neigh5.fit(X_traintrain,y_traintrain)
```
