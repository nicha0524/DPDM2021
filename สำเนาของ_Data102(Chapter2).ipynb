###Barchart

####stacked barchart
  https://matplotlib.org/stable/gallery/misc/table_demo.html#sphx-glr-gallery-misc-table-demo-py

```python
import numpy as np  #numpy : Number Python
import matplotlib.pyplot as plt
```

```python
# ข้อมูลภัยพิบัติ 
data = [[ 66386, 174296,  75131, 577908,  32015], # 5 year ย้อนหลัง
          [ 58230, 381139,  78045,  99308, 160454],     # 10year ย้อนหลัง
        [ 89135,  80552, 152558, 497981, 603535],    # 20 year ย้อนหลัง
        [ 78415,  81858, 150656, 193263,  69638],    # 50 year ย้อนหลัง
        [139361, 331509, 343164, 781380,  52269]]   # 100 year ย้อนหลัง
```

```python

columns = ('Freeze', 'Wind', 'Flood', 'Quake', 'Hail')
rows = ['%d year' % x for x in (100, 50, 20, 10, 5)] #%d : แสดงค่าจำนวนเต็ม 
# ['%d year' % x for x in (100, 50, 20, 10, 5)] เป็น format เก่า สามารถ ใช้ string แทนได้
```

```python
rows 
```

```python
values = np.arange(0, 2500, 500) # เริ่มจาก 0 เพิ่มขึ้นครั้งละ 500 
value_increment = 1000
```

```python
values
```

```python
value_increment
```

```python

# Get some pastel shades for the colors
colors = plt.cm.BuPu(np.linspace(0, 0.5, len(rows))) 
n_rows = len(data) #วัดค่าตาม data

index = np.arange(len(columns)) + 0.3 #index มีไว้ระบุตำแหน่ง วัดค่าตาม columns เริ่มจาก 0.3
bar_width = 0.4 # เป็นความกว้างของ barchart
```

```python
n_rows # จำนวน row = 5 วัดค่าตาม data
```

```python
index #index มีไว้ระบุตำแหน่ง เริ่มจาก 0.3 เพิ่มทีละ1
```

```python
colors # red green blue alpha(เป็นความโปร่งแสง จะทำให้เห็นค่าที่ซ่อนอยู่ได้)
```

```python
# Initialize the vertical-offset for the stacked bar chart.
y_offset = np.zeros(len(columns)) #offset คือ จำนวนตำแหน่งที่จะขยับไป โดยในที่นี้กำหนดให้เริ่มต้นคือ 0 ,1:ตำแหน่งปัจจุบัน, 2: ตำแหน่งสุดท้าย
```

```python
y_offset
```

```python
# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows): #วนลูป  row ทำซ้ำตามจำนวน n_rows
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])   # พล็อตกราฟ Barchart  x คือ index ,y คือ data[row] เรียงตามปี ,bar_width คือ ความกว้างของ barchart =0.4,bottom คือ จะบอกว่าให้แท่ากราฟไว้ตั้งที่ไหน 
    y_offset = y_offset + data[row] # ข้อมูลที่จะขยับไป+ข้อมูลใน Data ตาม row เช่นแท่งแรกตั้งที่ 0 + ข้อมูลData ใน Row แรก (5year) แท่งที่สองจะเป็น แท่ง+data ในแถวถัดไป
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset]) #เพิ่มข้อมูลลงใน cell_text โดยข้อมูลแต่ละตัวจะถูกหารด้วย1000 และบวกด้วย data[row] โดยใช้ลูปมาช่วยวน
```

```python
y_offset = np.zeros(len(columns))
# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1] 
cell_text.reverse() #reverse คือ การกลับตัวแห่ง
```

```python
y_offset = np.zeros(len(columns))
# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text, #สร้างตาราง ใช้ตำสั่ง plt.tabl โดย data คือ ข้อมูลในตารางคือ cell_text
                      rowLabels=rows,   #=ชื่อแถวตาม rows =['100 year', '50 year', '20 year', '10 year', '5 year']
                      rowColours=colors, # สีของ row ตาม colors
                      colLabels=columns, # ชื่อ column ตาม columns =('Freeze', 'Wind', 'Flood', 'Quake', 'Hail')
                      loc='bottom') #loc คือ การดึงข้อมูลจาก bottom
```

```python
y_offset = np.zeros(len(columns))
# Plot bars and create text labels for the table
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.2, bottom=0.2) 

plt.ylabel("Loss in ${0}'s".format(value_increment))      #ตั้งชื่อแกน y                
plt.yticks(values * value_increment, ['%d' % val for val in values]) #บอกตำแหน่งตามแกน y
plt.xticks([]) #บอกตำแหน่งตามแกน x
plt.title('Loss by Disaster') #ชื่อกราฟ

plt.show()
```
