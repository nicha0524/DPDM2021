###แบ่งกลุ่ม Profit  โดยใช้ scikit-learn (Clustering )

##### Standardize ข้อมูล

```python
from sklearn.preprocessing import StandardScaler
```

```python
Sales_OrderItems_std = pd.DataFrame(StandardScaler().fit_transform(Sales_OrderItems.iloc[:,[0,1,2,4,5]]))
```

```python
Sales_OrderItems_std 
```

```python
Sales_OrderItems_std =Sales_OrderItems_std.rename(columns = {0:'Sale_order_item',1:'Gross_Amount',2:'Net_Amount',3:'Prosit',4:'Name_product'})
```

```python
Sales_OrderItems_std
```

#####แบ่งข้อมูล

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=5,random_state=6)
Sales_OrderItems_std['cluster'] = cluster.fit_predict(Sales_OrderItems_std.iloc[:,[3]])
Sales_OrderItems_std.cluster.value_counts()
```

```python
Sales_OrderItems_std
```

```python
Sales_Order_Items=Sales_OrderItems_std.iloc[:,[0,1,2,4,5]]
```

```python
Sales_Order_Items =Sales_Order_Items.rename(columns = {'cluster':'Profit'})
```
