# Handling Missing Value 1 (ลบค่า missing )

````python
data_cases.shape #ดูขนาดของข้อมูล
````

````python
data_cases_M1 = data_cases.dropna() #.dropna() ตัดค่าผิดปกติ (ค่าผิดปกติในที่นี้คือค่าว่าง)
````

````python
data_cases_M1.shape
````

size before drop = 1000

size after drop = 10 

we lost 99% of data

````python
print('size before drop = ',len(data_cases))   #จำนวนข้อมูลก่อนตัด missing
print('size after drop = ', len(data_cases_M1)) #จำนวนข้อมูลหลังตัด missing
print('we lost ', ((len(data_cases)-len(data_cases_M1))/len(data_cases))*100, '% of data') #จำนวนข้อมูลหลังตัด missing
````

````python
# ใช้ f'string
print(f'size before drop = {len(data_cases)}') #จำนวนข้อมูลก่อนตัด missing 
print('size after drop = ', len(data_cases_M1)) #จำนวนข้อมูลหลังตัด missing
print(f'we lost {((len(data_cases)-len(data_cases_M1))/len(data_cases))*100:.2f}% of data')#จำนวนข้อมูลหลังตัด missing (: .2 กำหนดจุดทศนิยม)
````
