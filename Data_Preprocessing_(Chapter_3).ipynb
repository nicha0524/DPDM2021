# Handling Missing Value 1 (ลบค่า missing )

````python
data_cases.shape #ดูขนาดของข้อมูล
````

````python
data_cases_M1 = data_cases.dropna() #.dropna() ตัดค่าผิดปกติ (ค่าผิดปกติในที่นี้คือค่าว่าง)
````

````python
data_cases_M1.shape
````

size before drop = 1000

size after drop = 10 

we lost 99% of data

````python
print('size before drop = ',len(data_cases))   #จำนวนข้อมูลก่อนตัด missing
print('size after drop = ', len(data_cases_M1)) #จำนวนข้อมูลหลังตัด missing
print('we lost ', ((len(data_cases)-len(data_cases_M1))/len(data_cases))*100, '% of data') #จำนวนข้อมูลหลังตัด missing
````

````python
# ใช้ f'string
print(f'size before drop = {len(data_cases)}') #จำนวนข้อมูลก่อนตัด missing 
print('size after drop = ', len(data_cases_M1)) #จำนวนข้อมูลหลังตัด missing
print(f'we lost {((len(data_cases)-len(data_cases_M1))/len(data_cases))*100:.2f}% of data')#จำนวนข้อมูลหลังตัด missing (: .2 กำหนดจุดทศนิยม)
````

## Handling Missing Value 1.5 (ลบค่า missing เฉพาะใน column ที่เราสนใจ)

````python
data_cases_M15.shape #ดูขนาดของข้อมูลหลังจากที่ตัด missing แล้ว
````

````python
data_cases_M15
````

````python
A=(len(data_cases)/len(data_cases)) 
B1=(len(data_cases)-len(data_cases_M15))/len(data_cases) #data ตัด missing เฉพาะ Column ออกแล้ว
B2=(len(data_cases)-len(data_cases_M1))/len(data_cases)  # data ตัด missing ทั้งหมด ออกแล้ว

print(A)
print(B1)
print(B2)
````

````python
print(f'size before drop = {len(data_cases)}') #จำนวนข้อมูลก่อนตัด missing 
print('size after drop = ', len(data_cases_M15)) #จำนวนข้อมูลหลังตัด missing
print(f'we lost {((len(data_cases)-len(data_cases_M15))/len(data_cases))*100:.2f}% of data')#จำนวนข้อมูลหลังตัด missing (: .2 กำหนดจุดทศนิยม)
print(f'now we have {((A-B1)-(A-B2))*100:.2f}% more data than hondling_missing1' )
````

Handling Missing Value 3 แทนด้วยค่ากลาง 

````python
data_cases[['age']]=data_cases[['age']].fillna(data_cases[['age']].mean()) # เติม missing ด้วยค่าเฉลี่ย (.mean())
````

````python
data_cases.isnull().any() #ตรวจสอบค่า missing
````

````python
sorted(list(data_cases['sex'].dropna())) # เรียงข้อมูล sorted() ที่ตัด missing แล้ว .dropna()
````

````python
len(list(data_cases['sex'].dropna())) # ดูจำนวนข้อมูลที่ตัด missing แล้ว
````

````python
data_cases[['sex']].dropna().mode() # หาค่าฐานนิยม .mode() ในข้อมูลที่ตัด missing แล้ว 
````

````python
data_cases[['sex']].dropna().mode().iloc[0,0] #หา .mode()  #.iloc[0,0] ใช้เพื่อให้ข้อมูลที่เติมลงไปไม่เป็นตาราง(ชี้ข้อมุลใ column ที่ถูกเลือก)
````

````python
data_cases[['sex']]=data_cases[['sex']].fillna(data_cases[['sex']].dropna().mode().iloc[0,0]) #แทน .mode() ลงในข้อมูลที่เป็น missing #.iloc[0,0] ใช้เพื่อให้ข้อมูลที่เติมลงไปไม่เป็นตาราง(ชี้ข้อมุลใ column ที่ถูกเลือก)
````

````python
data_cases.isnull().any() #ตรวจสอบค่า missing ถ้าขึ้น False แสดงว่าไม่มี missing แล้ว
````

````python
data_cases[['province_of_onset']].dropna().mode().iloc[0,0] #หา .mode()  #.iloc[0,0] ใช้เพื่อให้ข้อมูลที่เติมลงไปไม่เป็นตาราง(ชี้ข้อมุลใ column ที่ถูกเลือก)
````

````python
data_cases[['province_of_onset']]=data_cases[['province_of_onset']].fillna(data_cases[['province_of_onset']].dropna().mode().iloc[0,0]) # แทน  .mode() ในค่าว่าง ของ column 'province_of_onset'
````
