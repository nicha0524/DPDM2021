###แก้ missing ใน province_of_isolation เติมค่าของ onset ใน isolation ก่อนจะเติมค่า mode

````python
#วน loop ตาม province_of_isolation แทนค่าในprovince_of_isolation ด้วย province_of_onset แต่ถ้า province_of_onset เป็นค่าว่างจะแทนค่าด้วย mode ของ province_of_isolation จังหวัดนั้น ๆ
new_p_iso = list()

for row in data_cases.iterrows():  #.iterrows()ซึ่งวนซ้ำบน DataFrame 
    if pd.isna(row[1]['province_of_isolation']): 
        if pd.isna(row[1]['province_of_onset']) == False:
            new_p_iso.append(row[1]['province_of_onset'])
        else:
            new_p_iso.append(data_cases[['province_of_isolation']].dropna().mode().iloc[0,0])
    else:
        new_p_iso.append(row[1]['province_of_isolation'])
````

````python
len(new_p_iso) # ดูความขนาดของ new_p_iso
````

````python
data_cases['province_of_isolation'] = new_p_iso #ให้ข้อมูล data_cases ใน column province_of_isolation เท่ากับ new_p_iso
````

#### Bin - Rep

#ตัดข้อมูล
* 0  :50     i=0   50*i : (50*i)+50
* 50 :100    i=1   50*i : (50*i)+50
* 100:150    i=2   50*i : (50*i)+50

````python
# วนลูปเลือกข้อมูล 10 bin ในแต่ละ bin มีสมาชิก 50 ตัว
import numpy as np

bin = 10
bin_r = list()
step = int(len(my_data_sorted)/bin)
for i in range(bin): 
    this_data = my_data_sorted[(step*i):(step*i)+step]    #ตัดข้อมูล
    rep = np.mean(this_data)   #เลือกตัวแทนแต่ละ bin (ค่าเฉลี่ยของชุดข้อมูล ใน bin นั้นๆ)
    bin_r.append(rep)
````

````python
bin_r # ค่าเฉลี่ย 10 bin
````

````python
list(range(0,len(my_data_sorted),50)) # ดู list ของ range
````

````python
# วนลูปเลือกข้อมูล 10 bin ในแต่ละ bin มีสมาชิก 50 ตัว
bin = 10
bin_r = list()
step = int(len(my_data_sorted)/bin) #step (ระยะห่าง)
for start in range(0,len(my_data_sorted),50):
    this_data = my_data_sorted[start:start+step]    #ตัดข้อมูล
    rep = np.mean(this_data)   #เลือกตัวแทนแต่ละ bin (ค่าเฉลี่ยของชุดข้อมูล ใน bin นั้นๆ)
    bin_r.append(rep)
````

````python
bin_r # ค่าเฉลี่ย 10 bin
````

####Smooth (การปรับเส้นให้เรียบ)

````python
my_data_smoothed = list() 
step = 0
this_bin = 0
for i in my_data_sorted: # วนลูปตาม my_data_sorted
    if step < 50:
        my_data_smoothed.append({'old_v':i,'new_v':bin_r[this_bin]}) # old_v (ค่าข้อมูลเดิม) คือ i ,  new_v ค่าข้อมูลใหม่ คือ bin_r[this_bin]
        step = step + 1
    else:
        step = 0
        this_bin = this_bin + 1
        my_data_smoothed.append({'old_v':i,'new_v':bin_r[this_bin]})
````

````python
my_data_smoothed # ข้อมูลที่ปรับแล้ว
````

````python
my_df = pd.DataFrame(my_data_smoothed) #แปลง Dict เป็น ตาราง pandas (ยังไม่ได้สอน)
my_df
````
